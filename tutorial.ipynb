```python
import numpy as np
import matplotlib.pyplot as plt
import librosa, librosa.display
%matplotlib inline
```

```python
# read file
file = r"F:\Audio\blues.00000.wav" # duration is 30s
```

```python
# waveform -> time domain

# signal will be a numpy array with sr * T elements
# 22050 * 30 
signal, sr = librosa.load(file, sr=22050) 
librosa.display.waveshow(signal, sr=sr)
plt.xlabel("Time") 
plt.ylabel("Amplitude")
plt.show()
```

```python
# fft -> frequency domain
fft = np.fft.fft(signal) # sr * T elements

magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))

plt.plot(frequency, magnitude)
plt.xlabel("frequency")
plt.ylabel("magnitude")
```

```python
left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude = magnitude[:int(len(magnitude)/2)]

plt.plot(left_frequency, left_magnitude)
plt.xlabel("frequency")
plt.ylabel("magnitude")
```

```python
# stft -> spectrogram

n_fft = 2048 # num times fft takes place AKA window
hop_length = 512
stft = librosa.core.stft(signal, n_fft=n_fft, hop_length=hop_length)

spectrogram = np.abs(stft)

librosa.display.specshow(spectrogram, sr=sr, hop_length=hop_length)
plt.xlabel("time")
plt.ylabel("frequency")
plt.colorbar()
```

```python
log_spectrogram = librosa.amplitude_to_db(spectrogram)
librosa.display.specshow(log_spectrogram, sr=sr, hop_length=hop_length)
plt.xlabel("time")
plt.ylabel("frequency")
plt.colorbar()
```

```python
# mffc

mffc = librosa.feature.mfcc(signal, n_fft=n_fft, hop_length=hop_length, n_mfcc=13)
librosa.display.specshow(mffc, sr=sr, hop_length=hop_length)
plt.xlabel("time")
plt.ylabel("mfcc")
plt.colorbar()
plt.show()
```
